```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
response = requests.get(URL)
soup = BeautifulSoup(response.content, 'html.parser')

first_team = soup.find("td", class_="rankings-block__banner--team-name").text.strip()
first_matches = soup.find("td", class_="rankings-block__banner--matches").text.strip()
first_points = soup.find("td", class_="rankings-block__banner--points").text.strip()
first_rating = soup.find("td", class_="rankings-block__banner--rating u-text-right").text.strip()

teams = [first_team]
matches = [first_matches]
points = [first_points]
ratings = [first_rating]

teams = soup.find_all('td', class_='table-body__cell rankings-table__team')
matches_points = soup.find_all('td', class_='table-body__cell u-center-text')
ratings = soup.find_all('td', class_='table-body__cell u-text-right rating')

team_names = [first_team] + [team.text.strip() for team in teams]
match_total = [first_matches] + [matches_points[i].text.strip() for i in range(0, len(matches_points), 2)]
point_total = [first_points] + [matches_points[i+1].text.strip() for i in range(0, len(matches_points), 2)]
rating_values = [first_rating] + [rating.text.strip() for rating in ratings]

df = pd.DataFrame({
    'Team': team_names,
    'Matches': match_total,
    'Points': point_total,
    'Rating': rating_values
})

print(df.head(10))

```

```python
Q7 Write a pythn program to scrape mentioned details from dineout.co.in and make dataframe-
1. Restaurant Name
2. Cuisine
3. Location
4. Ratings
5. Images URL
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = ('https://www.dineout.co.in')
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python


restaurant_names = []
cuisines = []
locations = []
ratings = []
image_urls = []

for restaurant in soup.find_all('div', class_="restaurant-card"):  
    name = restaurant.find('h4', class_="_1jbOb").text
    restaurant_names.append(name)

   
    location = restaurant.find('p', class_="_1jbOb").text
    locations.append(location)

  
    cuisine = restaurant.find('p', class_="cuisine-class").text
    cuisines.append(cuisine)

    
    rating = restaurant.find('div', class_="rating-class").text  # Adjust class name
    ratings.append(rating)

    
    image = restaurant.find('img')['src']
    image_urls.append(image)


df = pd.DataFrame({
    'Restaurant Name': restaurant_names,
    'Cuisine': cuisines,
    'Location': locations,
    'Ratings': ratings,
    'Image URL': image_urls
})



print(df)
```
