```python
Adjusted_grades = {'A+': 4.1, 'A': 4.0,'A-': 3.8, 'B': 3.1,'B+': 3.5, 'B-': 2.6,'C': 2.1,'C+': 2.5, 'C-': 1.8, 'D': 1.1, 'D+': 1.5, 'D-': 0.8, 'F': 0.0, 'Wu':0.0}
features = df.drop(["CS-403", "CS-421", "CS-406", "CS-414", "CS-419", "CS-423", "CS-412", 'MT-442', "Seat No.", "CGPA"], axis=1)
for f in features.columns:
    features[f] = features[f].map(Adjusted_grades)
target = df["CGPA"]
```

```python
from scipy.stats import boxcox
from sklearn.preprocessing import PowerTransformer
def reduce_skewness(column):
    skew = column.skew()
    if np.all(column >= 0):
        new = np.sqrt(column)
    else:
        transformer = PowerTransformer(method='yeo-johnson')
        new = transformer.fit_transform(column.values.reshape(-1, 1)).flatten()
    return new
features = features.apply(reduce_skewness)
```

```python
first_year_courses = ['PH-121', 'HS-101', 'CY-105', 'HS-105/12', 'MT-111', 'CS-105', 'CS-106', 'CS-107', 'EL-102', 'EE-119', 'ME-107']
second_year_courses = ['PH-121', 'HS-101', 'CY-105', 'HS-105/12', 'MT-111', 'CS-105', 'CS-106', 'EL-102', 'EE-119', 'ME-107', 'CS-107', 'HS-205/20', 'MT-222', 'EE-222', 'MT-224', 'CS-210', 'CS-211', 'CS-203', 'CS-214', 'EE-217', 'CS-212', 'CS-215']
third_year_courses = ['MT-331', 'EF-303', 'HS-304', 'CS-301', 'CS-302', 'TC-383', 'EL-332', 'CS-318', 'CS-306', 'CS-312', 'CS-317']
```

```python
def load_model(year):
    if year == 1:
        model = joblib.load('linear_regression_model_year1.joblib')
    elif year == 2:
        model = joblib.load('linear_regression_model_year2.joblib')
    elif year == 3:
        model = joblib.load('linear_regression_model_year3.joblib')
    else:
        raise ValueError("Invalid year. Please enter 1, 2, or 3.")
    return model
```

```python
def collect_grades(course_columns):
    user_grades = []
    for col in course_columns:
        student_grade_input = input(f"Enter your grade for {col}: ").strip().upper()
        numeric_grade = Adjusted_grades.get(student_grade_input, 0)  
        user_grades.append(numeric_grade)
    return user_grades

def predict_cgpa(year, grades):
    model =load_model(year)
    grades_array = np.array(grades).reshape(1, -1)
    predicted_cgpa = model.predict(grades_array)[0]
    return predicted_cgpa

student_name = input("Enter your name: ") 
year = int(input("Select the year for which you would like to predict your CGPA (1, 2, or 3): "))

if year == 1:
    grades = collect_grades(first_year)  
elif year == 2:
    grades = collect_grades(second_year)
elif year == 3:
    grades = collect_grades(third_year_combined)
else:
    raise ValueError("Invalid Year. Please enter 1, 2, or 3.")

predicted_cgpa = predict_cgpa(year, grades)
print(f"{student_name}, Estimated CGPA {year} is: {predicted_cgpa:.2f}")
```
