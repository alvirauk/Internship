```python
Q1. Write a python program which searches all the product under a particular product from www.amazon.in. The
product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for
guitars. 
In the above question, now scrape the following details of each product listed in first 3 pages of your search
results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then
scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 
```

```python
driver = webdriver.Chrome()
driver.get("https://www.amazon.in/s?k=samsung+flip+5")

products = []
product_class = "s-line-clamp-2" 

wait = WebDriverWait(driver, 10)

for i in range(2): 
    try:
       
        wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, product_class)))
        product_elements = driver.find_elements(By.CLASS_NAME, product_class)
        products.extend([p.text for p in product_elements if p.text != '']) 
    except TimeoutException:
        print("Timed out waiting for products to load.")
    
    try:
        next_button = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, 's-pagination-next')))
        next_button.click()
        time.sleep(2)  
    except TimeoutException:
        print("Timed out waiting for the next button to be clickable.")
        break
    except NoSuchElementException:
        print("Next button not found on the page.")
        break

print(f"Total products found: {len(products)}")
```
